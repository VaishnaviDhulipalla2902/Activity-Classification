```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import glob
```

```python
path = 'datafiles/'                    
all_files = glob.glob(os.path.join(path, "*.csv"))    

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
concatenated_df.head(-1)
```

```python
concatenated_df['label']
```

```python
var = 0
sub = 1
subject = []
for index, row in concatenated_df.iterrows():
    print(index)
    if( var == 100):
        var = 0
        sub = sub+1
    
    var+=1    
    subject.append(sub)
concatenated_df['subject'] = subject
        
```

```python
concatenated_df = concatenated_df.drop(['Milliseconds', 'Timestamp'], axis=1)

```

```python
concatenated_df.head()
```

```python
print(concatenated_df.to_string(index = 400))
```

```python
dataframe2 = concatenated_df.groupby('subject').agg(['mean', 'std'])
dataframe2.columns = ['X_acc_mean','X_acc_std', 'Y_acc_mean','Y_acc_std', 'Z_acc_mean','Z_acc_std', 'X_gyro_mean', 'X_gyro_std', 'Y_gyro_mean', 'Y_gyro_std', 'Z_gyro_mean', 'Z_gyro_std', 'label', 'l']
```

```python
dataframe2.head()
```

```python
dataframe2 = dataframe2.drop(['l'], axis = 1)
```

```python
dataframe2.head(-1)
```

```python
concatenated_df.head(-1)
```

```python


def RMS(x):
    return np.sqrt(sum(x**2/len(x)))

```

```python
dataframe3 = concatenated_df.groupby('subject').agg(['max', 'min', 'median', RMS])
dataframe3.head()
```

```python
dataframe3.columns = ['X-acc_max','X-acc_min','X-acc_median','X-acc_RMS', 'Y_acc_max','Y-acc_min','Y-acc_median','Y-acc_RMS', 'Z_acc_max','Z-acc_min','Z-acc_median','Z-acc_RMS', 'X_gyro_max','X-gyro_min','X-gyro_median','X-gyro_RMS', 'Y_gyro_max','Y_gyro_min','Y_gyro_median','Y_gyro_RMS', 'Z_gyro_max','Z_gyro_min','Z_gyro_median','Z_gyro_RMS', 'label1', 'label2', 'label3', 'label4']
dataframe3 = dataframe3.drop(['label1', 'label2', 'label3', 'label4'], axis=1)
```

```python
dataframe2.head()
```

```python
dataframe2 = pd.merge(dataframe2, dataframe3, on='subject')

```

```python
list(dataframe2.columns)
```

```python
dataframe2.head(-1)
```

```python
dataframe4 = concatenated_df.groupby(['subject']).agg(lambda x:x.value_counts().index[0])
dataframe4.columns = ['X_acc_mode', 'Y_acc_mode', 'Z_acc_mode', 'X_gyro_mode', 'Y_gyro_mode', 'Z_gyro_mode', 'label']
dataframe4 = dataframe4.drop(['label'], axis=1)
```

```python
dataframe4.head()
```

```python
dataframe2 = pd.merge(dataframe2, dataframe4, on='subject')
```

```python
dataframe2.head(-1)
```

```python
list(dataframe2.columns)
```

```python
train, test = train_test_split(dataframe2, test_size=0.2)
```

```python
Y_train = train['label']
X_train = train.drop(columns = ['label'])
```

```python
Y_test = test['label']
X_test = test.drop(columns = ['label'])
```

```python
X_train.head()
```

```python
# accuracy_scores = []
clf = SVC().fit(X_train, Y_train)
prediction = clf.predict(X_test)
accuracy_scores = accuracy_score(Y_test, prediction)*100
print('Support Vector Classifier accuracy: {}%'.format(accuracy_scores))
```
